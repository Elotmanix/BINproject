```python
# prompt: import matplotlib and matplot.pyplot and kmaen

import matplotlib as mlp
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
image=mlp.image.imread("/content/top-20-small-dog-breeds.jpeg.webp")
plt.imshow(image)
```

```python
image.shape
```

```python
X=image.reshape(-1,3)
```

```python
X.shape
```

```python
Kmeans=KMeans(n_clusters=5,n_init=10)
```

```python
Kmeans.fit(X)
```

```python
segmanted_image=Kmeans.cluster_centers_[Kmeans.labels_]
segmanted_image=segmanted_image.reshape(image.shape)
plt.imshow(segmanted_image/255)
```
